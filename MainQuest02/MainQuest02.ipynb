```python
import csv
import os
import random as r
os.chdir(r'C:\Users\SAMSUNG\OneDrive\바탕 화면\AIFFEL')

class Aiffel:
    def __init__ (self):
        self.Aiffel_list = {}
        self.core_num = sorted(core_list.keys())
        self.year = years%100
        self.gisu = gisu
        self.gwajung = gwajung
        self.scores = 0
        self.bonus_score = 0
        self.penalty_score = 0

        if self.gwajung == '코어':
            self.n_gwajung = 2
        else:
            self.n_gwajung = 1

        for n,name in enumerate(self.core_num):
            core_number = f"{self.year}{self.gisu:02}{self.n_gwajung}{n+1:02}"
            self.Aiffel_list[name] = [core_number, self.scores, self.bonus_score, self.penalty_score]

    def score(self):
        n = 0
        name = input("점수를 줄 학생의 이름을 입력하세요: ")
        while n<1 or n>3:
            n = int(input("줄 점수를 입력하세요: "))
            if n<1 or n>3 :
                print("올바르지 않은 점수입니다.")
            else:
                self.Aiffel_list[name][2] += n
                self.Aiffel_list[name][1] += self.Aiffel_list[name][2]

    def penalty(self):
        n = 0
        name = input("점수를 차감할 학생의 이름을 입력하세요: ")
        while n>-1:
            n = int(input("차감할 점수를 입력하세요: "))
            if n>-1 :
                print("올바르지 않은 점수입니다.")
            else:
                self.Aiffel_list[name][3] += n
                self.Aiffel_list[name][1] += self.Aiffel_list[name][3]

    def get_id_num(self):
        print(len(self.Aiffel_list))

    def display_info(self):
        with open('Aiffel.csv', 'w', newline='') as file:
            csvobject = csv.writer(file)
            csvobject.writerow
            csvobject.writerow(['이름', '학번', '과정종류', '퀘스트 점수 총점', '퀘스트 입력 총점', '퀘스트 페널티 총점'])
            for name in self.Aiffel_list:
                csvobject.writerow([name, self.Aiffel_list[name][0], self.gwajung, self.Aiffel_list[name][1], self.Aiffel_list[name][2], self.Aiffel_list[name][3]])

    def delete(self):
        name = input("중도포기한 학생의 이름을 입력하세요: ")
        del self.Aiffel_list[name]
        self.display_info()


class Grouping(Aiffel):
    def __init__(self):
        self.groups = {
            "상생": [], "신뢰": [], "열정": [], "이끔": [],
            "성장": [], "가치": [], "공유": [], "확산": []
        }

    def group(self):
        name = list(sorted(core_list.keys()))
        g_name = list(self.groups.keys())
        r.shuffle(g_name)
        r.shuffle(name)
        last = []
        while len(name)!=0 or len(g_name)!=0:
            rannum = r.randint(2,4)
            a = []
            if len(name)>rannum:
                for i in range(rannum):
                    a.append(name[0])
                    del name[0]
                self.groups[g_name[0]]
                del g_name[0]
            elif len(name) == 1:
                min_size = min(len(value) for value in self.groups.values())  # 각 리스트의 크기 중 가장 작은 값을 가져옵니다.
                for key, value in self.groups.items() :
                    if len(value) == min_size:
                        self.groups[key].append(name[0])
                        del name[0]
                        break
            else :
                continue

    def display_group(self):
        print(self.groups)


class Guilding(Aiffel):
    def __init__(self):
        self.guild = {
            "힌튼": [], "알트만": [], "응": [], "하사비스": []
        }




core_list = {'김건': 0,'김경민': 0, '김소연': 0, '김영신': 0, '박근수': 0, '박영수': 0, '박영준': 0, '서원영': 0, '신지만': 0, '심지안': 0, '이규상': 0, '이윤상': 0, '이진영': 0, '임현석': 0, '전요한': 0, '전재영': 0, '조대희':0}
years = 2023
gisu = 26
gwajung = '코어'
a = Grouping()
a.group()
a.display_group()
```
